```python
import time

size_of_vec = 10000000
def pure_python_version():
    t1 = time.time()
    X = range(size_of_vec)
    Y = range(size_of_vec)
    Z = []
    for i in range(len(X)):
        Z.append(X[i] + Y[i])
    return time.time() - t1

def numpy_version():
    t1 = time.time()
    X = np.arange(size_of_vec)
    Y = np.arange(size_of_vec)
    Z = X + Y
    return time.time() - t1
t1 = pure_python_version()
t2 = numpy_version()
print(t1, t2)
```

Plotting the sin function

```python
x = np.arange(-20,20,0.01)
y = np.sin(x)
plt.plot(x,y,'r')
plt.show()
```

Plotting the cos function

```python
z = np.cos(x)
plt.plot(x,z,'b')
plt.show()
```

Plotting in the same graph

```python
plt.plot(x,y,'r')
plt.plot(x,z,'b')
plt.show()
```

```python
t = np.tan(x)
plt.plot(x,t,'g')
plt.show()
```

Plotting with x square

```python
s = 0.01*x*x
plt.plot(x,y,'r')
plt.plot(x,z,'b')
plt.plot(x,s,'y')

plt.show()
```

Operations along an axis

```python
a = np.random.randint(0,9,size=(5,4))
print(a)
print(np.sum(a,axis=0)) #add the elements in vertical axis
print(np.sum(a,axis=1)) #add trhe elements in horizontal direction
```
