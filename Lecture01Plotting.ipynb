```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize empty lists to store data
x = []
y = []

# Read data from the file
with open('data01.txt', 'r') as file:
    for line in file:
        # Split the line into two values and convert them to floats
        values = line.split()
        x.append(float(values[0]))
        y.append(float(values[1]))

# Convert data to NumPy arrays for easier calculations
x = np.array(x)
y = np.array(y)

# Hyperparameters
learning_rate = 0.0001
epochs = 1000

# Initialize the parameters for the linear regression model
slope = 0
intercept = 0

# Perform gradient descent
for _ in range(epochs):
    # Predicted values
    y_pred = slope * x + intercept

    # Calculate the gradients
    d_slope = (-2 / len(x)) * np.sum(x * (y - y_pred))
    d_intercept = (-2 / len(x)) * np.sum(y - y_pred)

    # Update parameters
    slope -= learning_rate * d_slope
    intercept -= learning_rate * d_intercept

# Create the scatter plot
plt.scatter(x, y, label='Data Points')

# Plot the best-fit line
plt.plot(x, slope * x + intercept, color='red', label='Best Fit Line')

# Set labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Linear Regression with Gradient Descent')

# Show the plot
plt.legend()
plt.show()

# Print the best-fit line equation
print(f'Best-fit line: y = {slope:.2f}x + {intercept:.2f}')
```
